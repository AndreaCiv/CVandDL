```python
brand_name = "MONCLER"
path_folder_images = "/Users/andreacivitarese/PycharmProjects/CVandDL/dataset_detection/dataset/MONCLER"
index = 0
for file_name in sorted(os.listdir(path_folder_images)):
    if file_name == '.DS_Store':
        continue
    file_path = os.path.join(path_folder_images, file_name)
    img_data = cv2.imread(file_path)
    new_file_name = str(index) + '_' + brand_name + '.jpg'
    new_file_path = os.path.join(path_folder_images, new_file_name)
    cv2.imwrite(new_file_path, img_data)
    print(str(index))
    index += 1
```
